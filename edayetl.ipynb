Ahora, para la columna duration la separo para dejar por un lado los minutos que dura, y por otro una columna donde diga la cantidad de temporadas "seasons", y debajo se visualizan los dataframes para ver que todo vaya bien.

Lo siguiente es agregar una columna a cada dataframe con su respectiva plataforma, para luego concatenarlos y que cada Movie o Tv Show tenga a qué plataforma pertenece.

```python
netflixdf['platform'] = 'Netflix'
huludf['platform'] = 'Hulu'
disneydf['platform'] = 'Disney'
amazondf['platform'] = 'Amazon'
```

Ya que la columna duration quedó separada, se elimina porque no va a servir más.

```python
amazondf.drop(['duration'], axis=1)
huludf.drop(['duration'], axis=1)
disneydf.drop(['duration'], axis=1)
netflixdf.drop(['duration'], axis=1)
```

Se concatenan ahora los cuatro dataframes en uno solo, así realizar de forma más fácil las futuras consultas. Debajo vemos algunos ejemplos para ver que el dataframe haya quedado bien estructurado.

```python
df = pd.concat([amazondf, netflixdf, huludf, disneydf], axis=0)
df.sample(10)
```

```python
df.shape
```

Con respecto a las demás columnas, no voy a eliminarlas por el momento ya que pueden ser necesarias para futuras consultas.

Ahora se debe cambiar el tipo de dato de las columnas 'tiempo_dur' y 'seasons', a integer para que se puedan realizar las consultas.

```python
df['seasons'] = df['seasons'].astype(int)
df['min'] = df['min'].astype(int)
```

```python
df.dtypes
```

Por último, a los valores nulos se les completa con el string 'Sin dato', para que quede completo el dataframe.

```python
df.fillna('Sin dato', inplace=True)
df.sample(3) #Vemos 3 ejemplos
```

Al dataframe final, lo convertimos a un csv para trabajarlo en nuestro archivo main.py .


```python
df = pd.DataFrame(df)
df.to_csv("df.csv")
```
